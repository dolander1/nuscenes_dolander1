```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

## Load lattice
with open('data/sets/nuscenes-prediction-challenge-trajectory-sets/epsilon_8.pkl', 'rb') as f:
    trajectories8 = pickle.load(f)
lattice8 = np.array(trajectories8)
              
with open('data/sets/nuscenes-prediction-challenge-trajectory-sets/epsilon_4.pkl', 'rb') as f:
    trajectories4 = pickle.load(f)
lattice4 = np.array(trajectories4)
    
with open('data/sets/nuscenes-prediction-challenge-trajectory-sets/epsilon_2.pkl', 'rb') as f:
    trajectories2 = pickle.load(f)
lattice2 = np.array(trajectories2)

lattice_list = ["lattice8", "lattice4", "lattice2"]
modes_list = [np.arange(eval(lattice).shape[0]) for lattice in lattice_list]

plt.rcParams.update({'font.size': 14})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

### Plot all trajectories
fontsize = 16
for lattice, lattice_string, modes in zip(lattice_list, lattice_list, modes_list):
    # Enable LaTeX formatting
    [plt.plot(eval(lattice)[mode,:,0], eval(lattice)[mode,:,1]) for mode in modes]
    plt.title(fr'Trajectory set $\kappa_{lattice_string[-1]}$', fontsize=fontsize)
    plt.xlabel(r'$x$ (m)', fontsize=fontsize)
    plt.ylabel(r'$y$ (m)', fontsize=fontsize)
#     plt.legend()
    plt.savefig(f'saveResults/compareDatasets/compareLattices_eps={lattice_string}.pdf')
    plt.show()
```
